## Understanding Traces, State, and Events

Interactive visualizations are not just pictures — they are systems.

To understand how they work, separate three core concepts:

---

### 1. Trace (What Is Drawn)

A **trace** is one visual layer of data in a figure.

- In a scatter plot colored by `day`, each day is its own trace.
- Each trace contains:
  - Its own x values
  - Its own y values
  - Marker properties
  - Visibility settings

Think of a trace as one group of marks that can be updated independently.

---

### 2. Event (What the User Does)

An **event** is a user action.

Examples:
- Clicking a dropdown
- Dragging a selection box
- Moving a slider
- Clicking a legend item
- Zooming or panning

An event does not change the visualization directly.
It triggers something else to happen.

---

### 3. State (What Is Currently Selected or Active)

**State** is the stored result of an event.

Examples:
- `selected_day = "Sat"`
- `bin_count = 25`
- `selectedpoints = [3, 7, 14]`
- `scale = "log"`

State answers:
> What version of the visualization are we currently in?

---

## How They Work Together

Event  
→ State changes  
→ Trace updates  
→ Figure re-renders  

Example:

User selects "Sat" in dropdown  
→ Event fires  
→ `selected_day = "Sat"` (state)  
→ Only Saturday trace is shown  
→ Chart updates  

---

### Key Insight

- Traces define structure.
- Events trigger change.
- State stores configuration.
- Trace updates produce visual change.

Understanding this architecture makes debugging and designing interactive visualizations much easier.
